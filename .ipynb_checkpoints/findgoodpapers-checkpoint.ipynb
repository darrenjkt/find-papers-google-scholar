# 2. Sort all retrieved papers

```python
import pandas as pd
df = pd.read_csv('object_tracking.csv')
```

## Sort by year, citations

```python
df = df.sort_values(by=['Year','Citations'], ascending=[False, False])
df.head(100)
```

## Find top 2 papers for each year

```python
df1 = df.sort_values('Citations',ascending = False).groupby('Year').head(2)
df1 = df1.sort_values(by=['Year','cit/year'], ascending=[False, False])
df1.head(30)
```
